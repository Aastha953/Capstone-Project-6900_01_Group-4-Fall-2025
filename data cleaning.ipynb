```python
import pandas as pd
import os

# Read the tab-delimited text file
data = pd.read_csv("All Recorded PABT Bus.txt", delimiter="\t")

# Convert to datetime and keep only the date (remove time)
data['Start_Date'] = pd.to_datetime(data['Start_Date']).dt.date
data['End_Date'] = pd.to_datetime(data['End_Date']).dt.date

# Save cleaned CSV to Desktop
output_path = os.path.expanduser("~/Desktop/PABTBus.csv")
data.to_csv(output_path, index=False)

# Confirm save location
print("âœ… File successfully saved at:", output_path)
```

```python
import pandas as pd
import os

# Define file paths
desktop = os.path.expanduser("~/Desktop")
files = [
    os.path.join(desktop, "Traffic.csv"),
    os.path.join(desktop, "Speeds.csv"),
    os.path.join(desktop, "PABTPassenger.csv"),
    os.path.join(desktop, "PABTBus.csv")
]

# Read and combine all CSVs
dfs = []
for file in files:
    df = pd.read_csv(file)
    df['Source_File'] = os.path.basename(file)  # optional: add a column to identify the source
    dfs.append(df)

# Combine all into one dataframe
combined_df = pd.concat(dfs, ignore_index=True)

# Save combined CSV
output_path = os.path.join(desktop, "Combined_Traffic_Data.csv")
combined_df.to_csv(output_path, index=False)

# Confirm completion
print("âœ… All files successfully combined and saved at:", output_path)

# Optional: Display shape and preview
print("ðŸ“Š Combined data shape:", combined_df.shape)
print(combined_df.head())
```
